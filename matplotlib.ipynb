```python
import matplotlib.pyplot as plt
import numpy as np
```

Basic Line sample

```python
x = np.array([1, 2, 3, 4])

y = np.array([2, 4, 5, 8])

plt.plot(x, y)

plt.title("Basic line flot")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

Sample Line Informed

```python
x = np.array([1, 2, 3, 4])

#y = np.array([2, 4, 5, 8])
y = np.array([3, 5, 2, 8])


plt.plot( y, linestyle='dotted', color= "red", marker='o')

plt.title("Basic line flot")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("new Line")
plt.show()
```

Multiple Line

```python

x = np.array([2, 3, 4, 6])
y1 = np.array([2, 4, 5, 7])
y2 = np.array([3, 5, 6, 2])

plt.plot(x, y1, label="line 1", color='red', marker='s')
plt.plot(x, y2, label="line 2", color='purple', linestyle = "--", marker='o')
plt.title('multiple line styles' )
plt.xlabel('X value')
plt.ylabel('Y value')
plt.legend()
plt.show()
```

Bar Chart with Numpy and Maplotlib

```python
x = np.array(['Alfa', 'Betta', 'Cam', 'Del'])
y = np.array([4, 5, 3, 10])

plt.bar(x, y, color = 'orange')
plt.title('Bar Chart Ex')
plt.xlabel('Categories')
plt.ylabel('values')
plt.show()
```

## Scatter Plot with Numpy

```python
x = np.array([2, 3, 5, 6, 8])
y = np.array([4, 2, 5, 7, 8])

plt.scatter(x, y, color = 'blue')
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Histogram with Numpy and Matplotlib

```python
data = np.array([3, 4, 6, 2, 6, 7, 8, 10])

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram Ex')
plt.xlabel('sotuv')
plt.ylabel('narx')
plt.show()
```

Multiple **Subplot** Graphs

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([3, 6, 9, 12])

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')
plt.title("plot 1")

plt.subplot(1, 2, 2)
plt.bar(x, y2, color="blue")
plt.title('plot 2')
plt.tight_layout()
plt.show()
```
